```python
import matplotlib.pyplot as plt
import seaborn as sns


colunas_interessantes = ['track_popularity', 'danceability', 'energy', 'loudness', 'speechiness', 'acousticness', 'instrumentalness', 'liveness', 'tempo', 'duration_ms']

# Configurar o estilo dos gráficos usando Seaborn
sns.set(style="whitegrid")

# Histogramas
df[colunas_interessantes].hist(bins=20, figsize=(15, 10), layout=(3, 4), color='skyblue')
plt.suptitle('Histogramas das Variáveis', y=1.02)
plt.show()
```

```python
correlacoes = df[['danceability', 'energy', 'track_popularity','tempo']].corr()
print(correlacoes)

sns.heatmap(correlacoes, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Matriz de Correlação')
plt.show()
```

```python
df_copy = df.copy()
df_copy['track_album_release_date'] = pd.to_datetime(df_copy['track_album_release_date'], errors='coerce')

# Filtrar para datas entre 2015 e 2020
df_copy = df_copy[(df_copy['track_album_release_date'].dt.year >= 2015) & (df_copy['track_album_release_date'].dt.year <= 2020)]

# Agrupando pela data de lançamento do álbum e calculando a média da popularidade
tendencias_temporais = df_copy.groupby('track_album_release_date')['track_popularity'].mean()

# Plotar tendências temporais
plt.figure(figsize=(15, 6))
tendencias_temporais.plot(marker='o', linestyle='-', color='b')  # Pode ajustar a estilização do gráfico conforme necessário
plt.title('Tendência Temporal de Popularidade (2015-2020)')
plt.xlabel('Data de Lançamento do Álbum')
plt.ylabel('Média de Popularidade')
plt.grid(True)
plt.show()
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

features_para_agrupamento = ['danceability', 'energy']

df_numeric = df[features_para_agrupamento]

# Padronizar as features
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df_numeric)

# Aplicar KMeans
kmeans = KMeans(n_clusters=3, random_state=42)
df['cluster'] = kmeans.fit_predict(df_scaled)

# Visualizar os clusters
sns.scatterplot(x='danceability', y='energy', hue='cluster', data=df, palette='viridis')
plt.title('Agrupamento de Músicas por Danceability e Energy')
plt.show()
```

"cluster" refere-se ao grupo de pontos de dados que são semelhantes entre si em comparação com pontos de dados em outros clusters. O objetivo principal do agrupamento é dividir um conjunto de dados em grupos significativos, onde os pontos dentro de um mesmo grupo são mais semelhantes entre si do que com os pontos em outros grupos.

```python
# Análise de gêneros
df['playlist_genre'].value_counts().plot(kind='barh', color='#8A2BE2')
plt.title('Distribuição de Músicas por Gênero')
plt.show()

# Análise de subgêneros
df['playlist_subgenre'].value_counts().plot(kind='barh', color='lightcoral')
plt.title('Distribuição de Músicas por Subgênero')
plt.show()
```

```python
import re

def encontrar_artistas_com_caracteres_especiais(dataframe, coluna_artist):
    artistas_com_caracteres_especiais = dataframe[dataframe[coluna_artist].str.contains(r'[^\w\s]', regex=True, na=False)][coluna_artist].unique()


encontrar_artistas_com_caracteres_especiais(df, 'track_artist')
```

```python
def tratar_artistas_com_caracteres_especiais(dataframe, coluna_artist):
    # Substituir caracteres especiais por uma string vazia
    dataframe[coluna_artist] = dataframe[coluna_artist].str.replace(r'[^\w\s]', '', regex=True)


tratar_artistas_com_caracteres_especiais(df, 'track_artist')
```

```python
# Análise de características musicais
caracteristicas_artista = df.groupby('track_artist')[['track_popularity']].mean()

# Análise de popularidade média
popularidade_artista = df.groupby('track_artist')['track_popularity'].mean()

# Filtrar os 20 melhores artistas por popularidade média
top_20_artistas = popularidade_artista.nlargest(20)

# Filtrar as características apenas para os 20 melhores artistas
caracteristicas_top_20 = caracteristicas_artista.loc[top_20_artistas.index]

top_20_artistas.sort_values().plot(kind='barh', figsize=(15, 8), color='lightgreen')
plt.title('20 Melhores Artistas por popularidade')
plt.xlabel('Média de Popularidade')
plt.ylabel('Artistas')
plt.show()
```
